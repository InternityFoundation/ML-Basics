```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header=None)
```

```python
dataset
```

```python
transactions = []
for transaction in dataset:
    transactions.append([str(dataset[transaction][product]) for product in range(20)])
```

```python
transactions
```

```python
from apyori import apriori
rules = apriori(transactions, min_length=2, min_support=0.003, min_confidence=0.2, min_lift=3)
```
