# Import

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.colors import ListedColormap
```

```python
import warnings
warnings.filterwarnings("ignore")
```

# Preparing Data

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3,4]].values
```

# Using Dendogram

```python
import scipy.cluster.hierarchy as sch
```

```python
dendogram = sch.dendrogram(sch.linkage(X, method='ward'))
```

# Fitting Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
hi_cl = AgglomerativeClustering(n_clusters=5, linkage='ward', affinity='euclidean')
```

```python
y_hi_cl = hi_cl.fit_predict(X)
```

# Visualization

```python
color = ['red', 'orange', 'blue', 'yellow', 'violet']
for i in range(5):
    plt.scatter(X[y_hi_cl==i,0],X[y_hi_cl==i,1],s=100,c=color[i])
plt.show()
```
